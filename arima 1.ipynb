```python
from datetime import datetime, timedelta
import random

def generate_random_dates(start_date, end_date, num_dates):
    # Convert start_date and end_date strings to datetime objects
    start_datetime = datetime.strptime(start_date, '%Y-%m-%d')
    end_datetime = datetime.strptime(end_date, '%Y-%m-%d')

    # Calculate the total number of days in the date range
    total_days = (end_datetime - start_datetime).days

    # Generate random dates within the range
    random_dates = []
    for _ in range(num_dates):
        random_offset = random.randint(0, total_days)  # Random offset in days
        random_date = start_datetime + timedelta(days=random_offset)
        random_dates.append(random_date.strftime('%Y-%m-%d'))

    return random_dates

# Date range
start_date = '2024-02-01'
end_date = '2024-10-10'
num_dates = 100

random_dates = generate_random_dates(start_date, end_date, num_dates)
print(random_dates)
```

```python
import pandas as pd

df=pd.read_csv(r"D:\Users\kramanathan\Downloads\generated_values.csv")
print(df)
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
df['fin_trans_dt'] = pd.to_datetime(df['fin_trans_dt'])
result = df.groupby([df['fin_trans_dt'].dt.to_period('M'), 'fin_company_id']).agg({'fin_source_amt': 'sum', 'fin_company_id': 'count'}).rename(columns={'fin_source_amt': 'Sum_amt', 'fin_company_id': 'Count'}).reset_index()
```

```python
# Filter data for training (months 2 to 8)
train_data = result[(result['fin_trans_dt'].dt.month >= 2) & (result['fin_trans_dt'].dt.month <= 8)]

# Filter data for testing (months 8 and above)
test_data = result[~((result['fin_trans_dt'].dt.month >= 2) & (result['fin_trans_dt'].dt.month <= 8))]
```

```python
# Reset index of train_data and test_data
train_data.reset_index(drop=True, inplace=True)
test_data.reset_index(drop=True, inplace=True)
```

```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA


# Function to fit ARIMA model for each company
def fit_arima(df):
    # Initialize lists to store predictions
    count_predictions = []
    avg_sales_predictions = []

    # Loop over each company
    for company_id, company_data in df.groupby('fin_company_id'):
        # Fit ARIMA model for count
        count_model = ARIMA(company_data['Count'], order=(5,1,0))
        count_model_fit = count_model.fit()

        # Fit ARIMA model for avg_sales
        avg_sales_model = ARIMA(company_data['Sum_amt'], order=(5,1,0))
        avg_sales_model_fit = avg_sales_model.fit()

        # Predict next month's count and average sales
        count_pred = count_model_fit.forecast(steps=2)
        avg_sales_pred = avg_sales_model_fit.forecast(steps=2)

        # Append predictions to lists
        count_predictions.append((company_id, count_pred))
        avg_sales_predictions.append((company_id, avg_sales_pred))

    return count_predictions, avg_sales_predictions

# Train ARIMA models and make predictions on test data
count_preds, avg_sales_preds = fit_arima(train_data)


```
